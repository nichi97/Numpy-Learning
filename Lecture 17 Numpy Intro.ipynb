# Practice - Reversed Dictionary 

```python
def reverseDic(dic):
    dicReversed = {}
    for i in dic.items():
        dicReversed[i[1]]= i[0]
    return dicReversed
dic1 = {"a": "apple", "b":"banana", "c": "carrot"}
reverseDic(dic1)
```

## Concatenation

```python
grid = np.array([[1, 2, 3],
                 [4, 5, 6]])
```

```python
# Concatenate across the 0 direction
np.concatenate([grid, grid], axis = 0)
```

```python
# Concatenate across the 1 direction
np.concatenate([grid, grid], axis = 1)
```

```python
# What if you try to concatenate this array on a 1-d array? 
x = np.array([1,2,3]).reshape(1,3)

# What do you expect these to return? 
# np.concatenate([x,x], axis = 0)
# np.concatenate([x,x], axis = 1)
```

```python
np.concatenate([grid,x], axis = 1)
```

You can also use np.vstack and np.hstack functions

## Splitting

```python
x = np.array([1, 2, 3, 99, 99, 3, 2, 1]).reshape((1,-1))
x = np.concatenate([x,x], axis=0)
x
```

```python
x = np.array([1, 2, 3, 99, 99, 3, 2, 1]).reshape((1,-1))
x = np.concatenate([x,x], axis=0)

splitPt = np.arange(2,8,2)
rst = np.split(x, splitPt, axis=1)
rst[0][0,0] = 5
print(x)
```

```python
grid = np.arange(16).reshape((4, 4))
grid
```

```python
upper, lower = np.vsplit(grid, 2)
print(upper)
print(lower)
```

```python
left, right = np.hsplit(grid, [2])
print(left)
print(right)
```

```python
np.split(grid, [1], 0)
```
